```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- Carga de Datos ---
try:
    df_train = pd.read_csv('train.csv')
    df_test = pd.read_csv('test.csv')
    print("Archivos CSV cargados correctamente.")
except FileNotFoundError:
    print("Error: Asegúrate de que 'train.csv' y 'test.csv' estén en el directorio correcto.")
    exit()

# Establecer estilo visual para los gráficos
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Tamaño de figura por defecto

print("\n--- Fase 2: Análisis Exploratorio de Datos (EDA) ---")

print("\n=== 1. Inspección Inicial ===")

print("\nDimensiones de los DataFrames:")
print(f"Train: {df_train.shape}")
print(f"Test:  {df_test.shape}")

print("\nPrimeras 5 filas de df_train:")
print(df_train.head())

print("\nPrimeras 5 filas de df_test:")
print(df_test.head())

print("\nInformación general de df_train:")
df_train.info()

print("\nInformación general de df_test:")
df_test.info()

# Comprobación opcional de la secuencia de IDs
is_train_id_seq = df_train['id'].is_monotonic_increasing
is_test_id_seq = df_test['id'].is_monotonic_increasing
print(f"\n¿El 'id' en df_train es secuencial y creciente?: {is_train_id_seq}")
print(f"¿El 'id' en df_test es secuencial y creciente?: {is_test_id_seq}")

print("\n=== 2. Análisis de la Variable Objetivo ('fallo') ===")

fallo_counts = df_train['fallo'].value_counts()
fallo_perc = df_train['fallo'].value_counts(normalize=True) * 100

# Mostrar conteos y porcentajes de la variable objetivo
print("\nDistribución de la variable 'fallo':")
print(fallo_counts)
print(f"\nPorcentaje:\n{fallo_perc}")

plt.figure(figsize=(8, 5))
# Gráfico de barras: Distribución de la variable objetivo 'fallo'.
sns.countplot(x='fallo', data=df_train, palette='viridis')
plt.title('Distribución de la Variable Objetivo (fallo)')
plt.xlabel('Fallo (0: Normal, 1: Fallo)')
plt.ylabel('Cantidad')
plt.xticks(ticks=[0, 1], labels=['Normal (0)', 'Fallo (1)'])
total = len(df_train['fallo'])
# Añadir anotaciones de porcentaje y conteo a las barras
for i, count in enumerate(fallo_counts):
    plt.text(i, count + 50, f'{fallo_perc.iloc[i]:.2f}% ({count})', ha='center', va='bottom')
plt.ylim(0, max(fallo_counts)*1.15)
plt.show()

print("\n=== 3. Análisis Univariado ===")

print("\n--- 3.a Variables Numéricas ---")
numerical_cols = ['temperatura', 'presion', 'sensor_ruido', 'sensor_3']

# Mostrar estadísticas descriptivas para columnas numéricas
print("\nEstadísticas descriptivas (df_train):")
print(df_train[numerical_cols].describe())

# Preparar subplots para histogramas
n_cols = 2
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5))
axes = axes.flatten()

# Generar histogramas para cada variable numérica
for i, col in enumerate(numerical_cols):
    # Histograma: Distribución de la variable numérica.
    sns.histplot(df_train[col], kde=True, ax=axes[i], bins=30)
    axes[i].set_title(f'Distribución de {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frecuencia')

# Ocultar ejes vacíos si el número de variables es impar
if len(numerical_cols) < len(axes):
    for j in range(len(numerical_cols), len(axes)):
        fig.delaxes(axes[j])

plt.suptitle('Histogramas de Variables Numéricas (Train)', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# Preparar subplots para boxplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5))
axes = axes.flatten()

# Generar boxplots para cada variable numérica
for i, col in enumerate(numerical_cols):
    # Boxplot: Distribución y outliers de la variable numérica.
    sns.boxplot(y=df_train[col], ax=axes[i], palette='viridis')
    axes[i].set_title(f'Boxplot de {col}')
    axes[i].set_ylabel('')

# Ocultar ejes vacíos si el número de variables es impar
if len(numerical_cols) < len(axes):
    for j in range(len(numerical_cols), len(axes)):
        fig.delaxes(axes[j])

plt.suptitle('Boxplots de Variables Numéricas (Train)', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

print("\n--- 3.b Variables Categóricas ---")
categorical_cols = ['modo_operacion', 'operador']

# Generar gráficos de barras para cada variable categórica
for col in categorical_cols:
    counts = df_train[col].value_counts(dropna=False)
    perc = df_train[col].value_counts(dropna=False, normalize=True) * 100
    
    plt.figure(figsize=(10, 5))
    # Gráfico de barras horizontales: Distribución de la variable categórica.
    sns.countplot(y=col, data=df_train, order=counts.index, palette='viridis', hue=col, dodge=False, legend=False)
    plt.title(f'Distribución de {col}')
    plt.xlabel('Cantidad')
    plt.ylabel(col)
    # Añadir anotaciones de conteo y porcentaje a las barras
    for index, value in enumerate(counts):
        plt.text(value, index, f' {value} ({perc.iloc[index]:.2f}%)', va='center')
    plt.xlim(0, max(counts)*1.15)
    plt.show()

print("\n--- 3.c Variable Temporal (timestamp) ---")

# Convertir 'timestamp' a datetime, manejar errores coercitivamente
# Advertencia: Esto puede introducir NaNs si el formato es incorrecto
original_missing_train = df_train['timestamp'].isnull().sum()
original_missing_test = df_test['timestamp'].isnull().sum()
df_train['timestamp'] = pd.to_datetime(df_train['timestamp'], errors='coerce')
df_test['timestamp'] = pd.to_datetime(df_test['timestamp'], errors='coerce')
new_missing_train = df_train['timestamp'].isnull().sum()
new_missing_test = df_test['timestamp'].isnull().sum()

# Advertir si la conversión generó nuevos NaNs
if new_missing_train > original_missing_train:
    print(f"Advertencia: La conversión de 'timestamp' en df_train introdujo {new_missing_train - original_missing_train} valores nulos.")
if new_missing_test > original_missing_test:
     print(f"Advertencia: La conversión de 'timestamp' en df_test introdujo {new_missing_test - original_missing_test} valores nulos.")

# Extraer componentes de fecha/hora (asume que la conversión fue mayormente exitosa)
# Puede generar errores si la columna 'timestamp' es completamente NaT después de la conversión
df_train['year'] = df_train['timestamp'].dt.year
df_train['month'] = df_train['timestamp'].dt.month
df_train['day'] = df_train['timestamp'].dt.day
df_train['dayofweek'] = df_train['timestamp'].dt.dayofweek # Lunes=0, Domingo=6
df_train['hour'] = df_train['timestamp'].dt.hour

df_test['year'] = df_test['timestamp'].dt.year
df_test['month'] = df_test['timestamp'].dt.month
df_test['day'] = df_test['timestamp'].dt.day
df_test['dayofweek'] = df_test['timestamp'].dt.dayofweek
df_test['hour'] = df_test['timestamp'].dt.hour

# Mostrar rango de fechas (puede fallar si 'timestamp' es todo NaT)
print(f"\nRango de fechas en df_train: {df_train['timestamp'].min()} a {df_train['timestamp'].max()}")
print(f"Rango de fechas en df_test:  {df_test['timestamp'].min()} a {df_test['timestamp'].max()}")

print("\n=== 4. Análisis de Valores Faltantes ===")

# Mostrar conteo y porcentaje de valores faltantes en df_train
print("\nValores faltantes por columna (df_train):")
missing_train = df_train.isnull().sum()
missing_train_perc = (missing_train / len(df_train)) * 100
missing_train_df = pd.DataFrame({'Count': missing_train, 'Percentage': missing_train_perc})
print(missing_train_df[missing_train_df['Count'] > 0])

# Mostrar conteo y porcentaje de valores faltantes en df_test
print("\nValores faltantes por columna (df_test):")
missing_test = df_test.isnull().sum()
missing_test_perc = (missing_test / len(df_test)) * 100
missing_test_df = pd.DataFrame({'Count': missing_test, 'Percentage': missing_test_perc})
print(missing_test_df[missing_test_df['Count'] > 0])

print("\n--- Fin de los pasos iniciales del EDA (1-4) ---")

print("\n=== 5. Análisis Bivariado ===")

# Preparar subplots para boxplots comparativos (Numérico vs. Objetivo)
numerical_cols = ['temperatura', 'presion', 'sensor_ruido', 'sensor_3']
n_cols = 2
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 6))
axes = axes.flatten()

# Generar boxplots comparativos para cada variable numérica vs. 'fallo'
for i, col in enumerate(numerical_cols):
    # Boxplot comparativo: Variable numérica vs. Variable objetivo.
    sns.boxplot(x='fallo', y=col, data=df_train, ax=axes[i], palette='viridis', showfliers=True)
    axes[i].set_title(f'{col} vs. Fallo')
    axes[i].set_xlabel('Fallo (0: Normal, 1: Fallo)')
    axes[i].set_ylabel(col)

# Ocultar ejes vacíos
if len(numerical_cols) < len(axes):
    for j in range(len(numerical_cols), len(axes)):
        fig.delaxes(axes[j])

plt.suptitle('Variables Numéricas según la Variable Objetivo (Train)', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# Preparar datos temporales para análisis categórico (rellenar NaNs en operador)
categorical_cols = ['modo_operacion', 'operador']
df_train_temp = df_train.copy()
df_train_temp['operador'] = df_train_temp['operador'].fillna('Desconocido')

# Generar gráficos de barras de tasa de fallo para cada variable categórica
for col in categorical_cols:
    failure_rate = df_train_temp.groupby(col)['fallo'].mean().sort_values(ascending=False) * 100
    
    plt.figure(figsize=(10, 5))
    # Gráfico de barras: Tasa de fallo por categoría.
    sns.barplot(x=failure_rate.index, y=failure_rate.values, palette='viridis', order=failure_rate.index)
    plt.title(f'Tasa de Fallo (%) por {col}')
    plt.xlabel(col)
    plt.ylabel('Tasa de Fallo (%)')
    plt.xticks(rotation=45, ha='right')
    # Añadir anotaciones de porcentaje a las barras
    for index, value in enumerate(failure_rate.values):
        plt.text(index, value + 1, f'{value:.2f}%', ha='center', va='bottom')
    plt.ylim(0, max(failure_rate.values)*1.15)
    plt.show()

del df_train_temp # Limpiar dataframe temporal

# Analizar tasa de fallo vs. variables temporales, si existen
time_cols = ['year', 'month', 'dayofweek', 'hour']
time_cols = [col for col in time_cols if col in df_train.columns]

if time_cols:
    for col in time_cols:
        failure_rate_time = df_train.groupby(col)['fallo'].mean().sort_index() * 100

        plt.figure(figsize=(12, 5))
        if col in ['month', 'hour', 'year']:
             # Gráfico de línea: Tasa de fallo a lo largo del tiempo (mes, hora, año).
             sns.lineplot(x=failure_rate_time.index, y=failure_rate_time.values, marker='o')
             plt.ylabel('Tasa de Fallo (%)')
        elif col == 'dayofweek':
            day_names = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
            # Gráfico de barras: Tasa de fallo por día de la semana.
            sns.barplot(x=failure_rate_time.index, y=failure_rate_time.values, palette='viridis')
            plt.xticks(ticks=range(7), labels=day_names, rotation=45, ha='right')
            plt.ylabel('Tasa de Fallo (%)')
            # Añadir anotaciones de porcentaje a las barras
            for index, value in enumerate(failure_rate_time.values):
                plt.text(index, value + 0.5, f'{value:.2f}%', ha='center', va='bottom')
            plt.ylim(0, max(failure_rate_time.values)*1.15)

        plt.title(f'Tasa de Fallo (%) por {col}')
        plt.xlabel(col.capitalize())
        plt.grid(True)
        plt.show()
else:
    print("No se encontraron columnas temporales ('year', 'month', 'dayofweek', 'hour') para analizar.")

# Calcular y visualizar matriz de correlación numérica
correlation_matrix = df_train[numerical_cols].corr()
plt.figure(figsize=(8, 6))
# Mapa de calor: Correlación entre variables numéricas.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor de Correlación entre Variables Numéricas')
plt.show()

print("\n--- Fin del Análisis Bivariado ---")

print("\n--- Profundizando en EDA (Fase 2 Continuación) ---")

# Preparar dataframes con 'operador' rellenado para análisis posteriores
df_train_filled = df_train.copy()
df_train_filled['operador'] = df_train_filled['operador'].fillna('Desconocido')
df_test_filled = df_test.copy()
df_test_filled['operador'] = df_test_filled['operador'].fillna('Desconocido')

# Redefinir listas de columnas por si acaso
numerical_cols = ['temperatura', 'presion', 'sensor_ruido', 'sensor_3']
categorical_cols = ['modo_operacion', 'operador']
time_cols = ['year', 'month', 'dayofweek', 'hour']

# Generar scatter plot de Temperatura vs. Presión, coloreado por fallo
plt.figure(figsize=(10, 7))
# Gráfico de dispersión: Relación entre dos variables numéricas, coloreado por 'fallo'.
sns.scatterplot(x='temperatura', y='presion', data=df_train, hue='fallo', palette='viridis', alpha=0.6)
plt.title('Temperatura vs. Presión (Coloreado por Fallo)')
plt.xlabel('Temperatura')
plt.ylabel('Presión')
plt.legend(title='Fallo (0: Normal, 1: Fallo)')
plt.show()

# Generar pairplot (puede ser lento)
try:
    print("Generando pairplot (esto puede tardar)...")
    # Pairplot (matriz de gráficos): Relaciones bivariadas entre variables numéricas, coloreado por 'fallo'.
    sns.pairplot(df_train, vars=numerical_cols, hue='fallo', palette='viridis', plot_kws={'alpha': 0.5})
    plt.suptitle('Pairplot de Variables Numéricas por Fallo', y=1.02)
    plt.show()
except Exception as e:
    print(f"No se pudo generar el pairplot completo: {e}. Considera usar una muestra.")

# Generar boxplots facetados de Presión vs. Fallo por Modo de Operación
# Catplot (Boxplots facetados): Presión vs. Fallo, separado por Modo de Operación.
sns.catplot(x='fallo', y='presion', col='modo_operacion', data=df_train, kind='box', palette='viridis', showfliers=True)
plt.suptitle('Presión vs. Fallo por Modo de Operación', y=1.03)
plt.xlabel('Fallo (0: Normal, 1: Fallo)') # Ajustar etiqueta X global
plt.show()

# Analizar tasa de fallo horaria (laborable vs. fin de semana), si existen columnas
if 'dayofweek' in df_train.columns and 'hour' in df_train.columns:
    df_train['EsFinDeSemana'] = df_train['dayofweek'].isin([5, 6])
    failure_rate_hourly_weekday = df_train.groupby(['EsFinDeSemana', 'hour'])['fallo'].mean().unstack(level=0) * 100
    failure_rate_hourly_weekday.columns = ['Laborable', 'FinDeSemana']

    plt.figure(figsize=(12, 6))
    # Gráfico de línea múltiple: Tasa de fallo por hora, comparando días laborables vs. fines de semana.
    sns.lineplot(data=failure_rate_hourly_weekday, markers=True)
    plt.title('Tasa de Fallo (%) por Hora del Día')
    plt.xlabel('Hora del Día')
    plt.ylabel('Tasa de Fallo (%)')
    plt.xticks(range(0, 24))
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.legend(title='Tipo de Día')
    plt.show()
else:
    print("No se pueden generar gráficos horarios por día laborable/fin de semana (faltan columnas 'dayofweek' o 'hour').")

# Analizar número de fallos a lo largo del tiempo (mensual), si existe timestamp
if 'timestamp' in df_train.columns:
    df_train_sorted = df_train.sort_values('timestamp').copy()
    # Asume que 'timestamp' no es completamente NaT
    df_train_sorted['YearMonth'] = df_train_sorted['timestamp'].dt.to_period('M')
    fallos_por_mes = df_train_sorted.groupby('YearMonth')['fallo'].sum()
    fallos_por_mes.index = fallos_por_mes.index.to_timestamp()

    plt.figure(figsize=(15, 6))
    # Gráfico de línea: Evolución temporal del número de fallos mensuales.
    fallos_por_mes.plot(kind='line', marker='o', linestyle='-')
    plt.title('Número de Fallos Registrados por Mes')
    plt.xlabel('Fecha (Mes)')
    plt.ylabel('Número Total de Fallos')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()
else:
    print("No se puede generar gráfico de fallos por mes (falta columna 'timestamp').")

# Comparar distribuciones numéricas entre Train y Test
n_cols_plot = 2
n_rows_plot = (len(numerical_cols) + n_cols_plot - 1) // n_cols_plot
fig, axes = plt.subplots(n_rows_plot, n_cols_plot, figsize=(15, n_rows_plot * 5))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    # Gráfico de densidad (KDE): Comparación de distribuciones de variables numéricas (Train vs. Test).
    sns.kdeplot(df_train[col].dropna(), ax=axes[i], label='Train', color='blue', fill=True, alpha=0.5) # dropna agregado por seguridad
    sns.kdeplot(df_test[col].dropna(), ax=axes[i], label='Test', color='red', fill=True, alpha=0.5) # dropna agregado por seguridad
    axes[i].set_title(f'Distribución de {col}: Train vs Test')
    axes[i].set_xlabel('')
    axes[i].legend()

# Ocultar ejes vacíos
if len(numerical_cols) < len(axes):
    for j in range(len(numerical_cols), len(axes)):
        fig.delaxes(axes[j])

plt.suptitle('Comparación de Distribuciones Numéricas: Train vs Test', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# Comparar distribuciones categóricas entre Train y Test
for col in categorical_cols:
    train_counts = df_train_filled[col].value_counts(normalize=True).rename('Train')
    test_counts = df_test_filled[col].value_counts(normalize=True).rename('Test')
    df_compare = pd.concat([train_counts, test_counts], axis=1).fillna(0) * 100

    # Gráfico de barras comparativo: Proporción de categorías (Train vs. Test).
    df_compare.plot(kind='bar', figsize=(10, 5))
    plt.title(f'Comparación de Proporciones de {col}: Train vs Test')
    plt.ylabel('Porcentaje (%)')
    plt.xlabel(col)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Conjunto')
    plt.grid(axis='y', linestyle='--', linewidth=0.5)
    plt.show()

# Analizar hipótesis: Sensores durante modo 'manual' vs. otros
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
# Boxplot comparativo: Variable numérica por categoría.
sns.boxplot(x='modo_operacion', y='temperatura', data=df_train, palette='viridis', order=['manual','auto','mantenimiento'])
plt.title('Temperatura por Modo Operación')
plt.subplot(1, 2, 2)
# Boxplot comparativo: Variable numérica por categoría.
sns.boxplot(x='modo_operacion', y='presion', data=df_train, palette='viridis', order=['manual','auto','mantenimiento'])
plt.title('Presión por Modo Operación')
plt.tight_layout()
plt.show()

# Analizar hipótesis: Sensores Operador 'E' vs. 'B'
df_operador_compare = df_train_filled[df_train_filled['operador'].isin(['E', 'B'])]

if not df_operador_compare.empty:
    n_cols_plot = 2
    n_rows_plot = (len(numerical_cols) + n_cols_plot - 1) // n_cols_plot
    fig, axes = plt.subplots(n_rows_plot, n_cols_plot, figsize=(15, n_rows_plot * 5))
    axes = axes.flatten()

    for i, col in enumerate(numerical_cols):
        # Boxplot comparativo: Variable numérica por categoría (Operador E vs B).
        sns.boxplot(x='operador', y=col, data=df_operador_compare, ax=axes[i], palette='viridis')
        axes[i].set_title(f'{col}: Operador E vs B')
        axes[i].set_xlabel('Operador')
        axes[i].set_ylabel(col)

    # Ocultar ejes vacíos
    if len(numerical_cols) < len(axes):
        for j in range(len(numerical_cols), len(axes)):
            fig.delaxes(axes[j])

    plt.suptitle('Comparación de Sensores: Operador E vs Operador B', y=1.02, fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.98])
    plt.show()
else:
    print("No hay datos suficientes de los operadores 'E' y 'B' para comparar.")

# Limpieza de DataFrames temporales
del df_train_filled
del df_test_filled
if 'df_train_sorted' in locals(): del df_train_sorted
if 'df_operador_compare' in locals(): del df_operador_compare
if 'failure_rate_hourly_weekday' in locals(): del failure_rate_hourly_weekday
if 'fallos_por_mes' in locals(): del fallos_por_mes
if 'df_compare' in locals(): del df_compare
print("\nDataFrames temporales eliminados.")

print("\n--- Fin de Análisis EDA Adicional ---")
```

```python
'''
import pandas as pd
import numpy as np

# Preprocessing
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer # Para manejar NaNs numéricos si los hubiera

# Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB # Requiere datos densos (OHE está bien)
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
# Nota: Para Random Forest y Gradient Boosting, librerías como XGBoost o LightGBM
# suelen ofrecer mejor rendimiento y más control, pero empezamos con scikit-learn.

# Metrics
from sklearn.metrics import f1_score, make_scorer

# --- 1. Carga de Datos ---
try:
    df_train = pd.read_csv('train.csv')
except FileNotFoundError:
    print("Error: Asegúrate de que 'train.csv' esté en el directorio correcto.")
    exit()

print("Datos cargados correctamente.")
print(f"Forma inicial del DataFrame: {df_train.shape}")
print("Primeras filas:")
print(df_train.head())
print("\nInformación del DataFrame:")
df_train.info() # Útil para ver tipos y nulos iniciales

# --- 2. Definición de Variables ---
target = 'fallo'

# Identifica columnas numéricas y categóricas (ajusta según tu CSV real)
# Asumimos que las columnas temporales (hora, día, mes) NO están aún creadas.
# Si tienes una columna de timestamp, deberías extraerlas primero.
numerical_features_base = ['temperatura', 'presion', 'sensor_ruido', 'sensor_3']
categorical_features_base = ['modo_operacion', 'operador']

# Verifica que las columnas existen en el DataFrame
actual_numerical_features = [col for col in numerical_features_base if col in df_train.columns]
actual_categorical_features = [col for col in categorical_features_base if col in df_train.columns]

if len(actual_numerical_features) != len(numerical_features_base):
    print("Advertencia: No todas las características numéricas esperadas se encontraron.")
if len(actual_categorical_features) != len(categorical_features_base):
    print("Advertencia: No todas las características categóricas esperadas se encontraron.")
if target not in df_train.columns:
    print(f"Error: La columna objetivo '{target}' no se encuentra en el DataFrame.")
    exit()

print(f"\nCaracterísticas numéricas identificadas: {actual_numerical_features}")
print(f"Características categóricas identificadas: {actual_categorical_features}")
print(f"Variable objetivo: {target}")

# --- 3. Preprocesamiento Básico Inicial ---
# Manejo específico de NaNs en 'operador' (según análisis previo)
if 'operador' in actual_categorical_features:
    # Crea una copia para no modificar el original directamente en cada iteración
    df_processed = df_train.copy()
    df_processed['operador'] = df_processed['operador'].fillna('Desconocido')
    print("\nValores NaN en 'operador' reemplazados por 'Desconocido'.")
    # Verifica si hay NaNs en otras columnas
    print("\nComprobación de NaNs restantes:")
    print(df_processed.isnull().sum())
else:
    df_processed = df_train.copy()

# Separar X e y
X = df_processed.drop(target, axis=1)
y = df_processed[target]

# --- 4. Definición de Pipelines de Preprocesamiento ---

# Pipeline para variables numéricas:
# - Imputar NaNs (si los hubiera) con la mediana (más robusto a outliers)
# - Escalar los datos
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Pipeline para variables categóricas:
# - Imputar NaNs (ya hecho para 'operador', pero por si acaso en otras) con una constante
# - Aplicar One-Hot Encoding
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')), # Por si hay otros NaNs categóricos
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # handle_unknown para robustez en CV
])

# --- 5. Definición de Conjuntos de Características (Feature Sets) ---

feature_sets = {}

# Set 1: Todas las variables originales
preprocessor_all = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, actual_numerical_features),
        ('cat', categorical_transformer, actual_categorical_features)
    ],
    remainder='drop' # No incluir otras columnas si las hubiera
)
feature_sets['Todas_Originales'] = {
    'features': actual_numerical_features + actual_categorical_features,
    'preprocessor': preprocessor_all
}
print("\nDefinido Set 1: Todas las características originales")

# Set 2: Variables más prometedoras (Quitamos 'sensor_ruido')
promising_numerical = ['temperatura', 'presion', 'sensor_3']
promising_categorical = ['modo_operacion', 'operador'] # Mantenemos las categóricas prometedoras
# Asegurarse que existen
promising_numerical = [f for f in promising_numerical if f in actual_numerical_features]
promising_categorical = [f for f in promising_categorical if f in actual_categorical_features]

if 'sensor_ruido' in promising_numerical: # Asegurarse de quitarlo si estaba en la lista base
    promising_numerical.remove('sensor_ruido')

preprocessor_promising = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, promising_numerical),
        ('cat', categorical_transformer, promising_categorical)
    ],
    remainder='drop'
)
feature_sets['Prometedoras'] = {
    'features': promising_numerical + promising_categorical,
    'preprocessor': preprocessor_promising
}
print("Definido Set 2: Características prometedoras")

# Set 3: Todas las originales + Interacciones (Polinómicas de grado 2 solo entre numéricas)
# Creamos un pipeline que primero preprocesa y luego añade interacciones

# Preprocesador solo para obtener las numéricas escaladas y las categóricas codificadas
preprocessor_base_interactions = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, actual_numerical_features),
        ('cat', categorical_transformer, actual_categorical_features)
    ],
    remainder='passthrough' # Mantenemos otras columnas (aunque no debería haber)
)

# Necesitamos saber cuántas columnas numéricas escaladas hay para aplicar PolyFeatures
# Esto es un poco más complejo dentro de un pipeline general.
# Una forma es aplicar PolyFeatures *después* del ColumnTransformer completo.
# Advertencia: Esto creará interacciones entre dummies categóricas, lo cual no siempre es deseado.
# Una aproximación más controlada sería crear un pipeline específico.

# Alternativa más simple: Crear un pipeline que incluya PolyFeatures al final
# Aplicará PolyFeatures a TODO lo que salga del ColumnTransformer (numéricas escaladas + categóricas OHE)
pipeline_with_poly = Pipeline([
    ('preprocess', preprocessor_all), # Usa el preprocesador del Set 1
    ('poly', PolynomialFeatures(degree=2, interaction_only=False, include_bias=False))
    # interaction_only=False crea x^2, y^2, x*y
    # include_bias=False evita añadir una columna de unos
])
# Para este set, el 'preprocessor' es todo el pipeline hasta PolyFeatures
feature_sets['Todas_Mas_Interacciones'] = {
    'features': actual_numerical_features + actual_categorical_features, # Features originales que entran
    'preprocessor': pipeline_with_poly # El preprocesador AHORA incluye PolyFeatures
}
print("Definido Set 3: Todas las características + Interacciones Polinómicas (Aplicado a todo post-procesamiento inicial)")
print("Nota: El Set 3 crea interacciones entre *todas* las variables preprocesadas (incluyendo dummies).")


# --- 6. Definición de Modelos a Probar ---
models = {
    "DecisionTree": DecisionTreeClassifier(random_state=42, class_weight='balanced'), # class_weight para desbalanceo
    "GaussianNB": GaussianNB(), # No tiene class_weight, más sensible a desbalanceo
    "RandomForest": RandomForestClassifier(random_state=42, n_estimators=100, class_weight='balanced', n_jobs=-1), # n_jobs=-1 usa todos los cores
    "GradientBoosting": GradientBoostingClassifier(random_state=42, n_estimators=100) # GB de sklearn no tiene class_weight directo
    # Podrías añadir aquí:
    # "XGBoost": xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', scale_pos_weight=ratio_neg_pos) # scale_pos_weight para desbalanceo
    # "LightGBM": lgb.LGBMClassifier(random_state=42, class_weight='balanced')
}

# --- 7. Configuración de Validación Cruzada ---
n_splits = 5 # Número de folds
cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
# Usamos 'f1' como métrica. Para binario es f1 por defecto de la clase positiva (1)
# Si fuera multiclase o necesitaras otra media, usarías make_scorer(f1_score, average='weighted') por ejemplo.
scoring_metric = 'f1'

# --- 8. Bucle de Experimentación ---
results = []

print(f"\n--- Iniciando Experimentación con {n_splits}-Fold Cross-Validation ---")
print(f"Métrica de evaluación: {scoring_metric}")

for set_name, set_config in feature_sets.items():
    print(f"\nEvaluando Feature Set: {set_name}")
    current_preprocessor = set_config['preprocessor']
    # Seleccionamos solo las columnas necesarias para este set ANTES de pasarlo al pipeline
    current_X = X[set_config['features']].copy()

    for model_name, model in models.items():
        # Crear el pipeline completo: Preprocesador (puede incluir PolyFeatures) + Modelo
        # Si el preprocesador ya es un pipeline (como en Set 3), esto anida pipelines
        if isinstance(current_preprocessor, Pipeline): # Caso Set 3
             # El modelo se añade como último paso al pipeline existente
             full_pipeline = Pipeline(steps=current_preprocessor.steps + [('model', model)])
        else: # Casos Set 1 y Set 2
             # El preprocesador es un ColumnTransformer
             full_pipeline = Pipeline(steps=[('preprocess', current_preprocessor),
                                            ('model', model)])

        try:
            # Realizar Validación Cruzada
            scores = cross_val_score(full_pipeline, current_X, y,
                                     cv=cv, scoring=scoring_metric, n_jobs=-1) # n_jobs=-1 usa todos los cores

            results.append({
                'Feature Set': set_name,
                'Model': model_name,
                f'Mean {scoring_metric.upper()}': np.mean(scores),
                f'Std Dev {scoring_metric.upper()}': np.std(scores),
                'Scores per Fold': scores
            })
            print(f"  {model_name}: Mean F1 = {np.mean(scores):.4f} (+/- {np.std(scores):.4f})")

        except Exception as e:
            print(f"  ERROR ejecutando {model_name} con {set_name}: {e}")
            results.append({
                'Feature Set': set_name,
                'Model': model_name,
                f'Mean {scoring_metric.upper()}': np.nan,
                f'Std Dev {scoring_metric.upper()}': np.nan,
                'Scores per Fold': [np.nan] * n_splits
            })

# --- 9. Presentación de Resultados ---
results_df = pd.DataFrame(results)
print("\n--- Resumen de Resultados ---")
# Ordenar por la métrica media descendente para ver los mejores primero
results_df = results_df.sort_values(by=f'Mean {scoring_metric.upper()}', ascending=False)
print(results_df[[ 'Feature Set', 'Model', f'Mean {scoring_metric.upper()}', f'Std Dev {scoring_metric.upper()}']])

print("\n--- Fin de la Experimentación ---")

# --- Consideraciones Adicionales / Próximos Pasos ---
print("\nConsideraciones y Próximos Pasos:")
print("1.  **Ingeniería de Características Temporales:** Añadir variables como hora, día de la semana, mes, día laborable/fin de semana (muy prometedoras según el análisis EDA).")
print("2.  **Hyperparameter Tuning:** Optimizar los parámetros de los mejores modelos (RandomForest, GradientBoosting) usando GridSearchCV o RandomizedSearchCV.")
print("3.  **Modelos Avanzados:** Probar XGBoost, LightGBM o CatBoost, que suelen superar a las implementaciones de scikit-learn y tienen mejor manejo de desbalanceo/categorías.")
print("4.  **Manejo de Desbalanceo:** Se usó 'class_weight', pero explorar otras técnicas como SMOTE (oversampling) o NearMiss (undersampling) podría ser beneficioso (usar `imbalanced-learn` library).")
print("5.  **Análisis de Errores:** Investigar dónde fallan los mejores modelos (matriz de confusión, análisis de predicciones incorrectas).")
print("6.  **Interacciones (Set 3):** Revisar si las interacciones polinómicas realmente aportaron valor o si introdujeron demasiado ruido/complejidad. Considerar interacciones más específicas basadas en el EDA.")
```
